```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

CSV = 'raw/D0_swarmplot.csv'
```

```python
df = pd.read_csv(CSV)
df.set_index('Gene', inplace=True)
display(df)
```

```python
df = df.loc[(df['Gene Type'] == 'Zf') | (df['Gene Type'] == 'bHLH') | (df['Gene Type'] == 'Homeobox') | (df['Gene Type'] == 'HTH') | (df['Gene Type'] == 'bZIP') | (df['Gene Type'] == 'HMG') | (df['Gene Type'] == 'CCAAT')]
df = df.loc[df['Log P-value'] >= 500]
display(df)
```

```python
ax = sns.swarmplot(x='Gene Type', y='Log P-value', hue='Gene Type', data=df, size=6, dodge=False)
figure = plt.gcf()
figure.set_size_inches(12,3.5)
plt.savefig('out/swarmplot.pdf', dpi=400)
```
