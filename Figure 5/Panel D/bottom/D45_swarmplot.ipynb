```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv('swarmplots/0_500cutoff_perc.csv')
df.set_index('Gene', inplace=True)
display(df)
```

```python
# df = df.loc[(df['Gene Type'] == 'Zf') | (df['Gene Type'] == 'bHLH') | (df['Gene Type'] == 'Homeobox')]
# df

df = df.loc[df['Log P-value'] >= 500]
```

```python
ax = sns.swarmplot(x='Gene Type', y='Log P-value', hue='Gene Type', data=df, size=6, dodge=False)
# sns.swarmpl(x='Gene Type', y='-logp', data=df)
figure = plt.gcf()
figure.set_size_inches(12,3.5)
plt.savefig('0_swarmplot_cutoff_500_SHORT.pdf', dpi=400)
```
